# Definiendo valores de ecut y phonones

```python
ecuts = ['36_0','40_0','44_0','48_0','52_0']
phonons = range(1,7)
```

# Función principal

```python
data = list()

for ecut in ecuts:
    os.chdir(ecut)
    ecut = float(ecut.replace('_','.'))
    for phonon in phonons:
        with open(f'x000.dyn{phonon}') as dyn_file:
            for line in dyn_file:
                if re.search('freq',line):
                    freq = re.findall('freq\s\(\s{1,4}\d\)',line)
                    freq = int(freq[0][9:-1])
                    freq_value = re.findall('\d{1,2}\.\d{1,6}\s\[THz\]',line)
                    freq_value = float(freq_value[0][:-6])
                    data.append([ecut, phonon, freq,freq_value])
    os.chdir('..')
```

```python
df = pd.DataFrame(data, columns=['ecut','phonon','freq','value'])
```

```python
for phonon in phonons:
    freq = df[(df['phonon']== phonon)].sort_values(by=['freq','ecut'])
    freq_1 = freq.iloc[0:5,:]
    freq_2 = freq.iloc[5:10,:]
    freq_3 = freq.iloc[10:15,:]
    freq_4 = freq.iloc[15:20,:]
    freq_5 = freq.iloc[20:25,:]
    freq_6 = freq.iloc[25:30,:]
    freq_7 = freq.iloc[30:35,:]
    freq_8 = freq.iloc[35:40,:]
    freq_9 = freq.iloc[40:45,:]

    fig, ax = plt.subplots()
    fig.set_size_inches(10,7)
    ax.plot(freq_1['ecut'],freq_1['value'])
    ax.plot(freq_2['ecut'],freq_2['value'])
    ax.plot(freq_3['ecut'],freq_3['value'])
    ax.plot(freq_4['ecut'],freq_4['value'])
    ax.plot(freq_5['ecut'],freq_5['value'])
    ax.plot(freq_6['ecut'],freq_6['value'])
    ax.plot(freq_7['ecut'],freq_7['value'])
    ax.plot(freq_8['ecut'],freq_8['value'])
    ax.plot(freq_9['ecut'],freq_9['value'])
    plt.xticks([36,40,44,48,52])
    plt.xlabel('Ecut',fontsize=14)
    plt.ylabel('Frecuencia [THz]',fontsize=14)
    plt.title(f'Phonon {phonon}', fontsize = 14)
    plt.savefig(f'/home/marcocarmonaga/Documents/MoB_2-Study-Properties_Quantum_expresso/images/frecuencias_grouped/Phonon_{phonon}.png',facecolor='white')
```
